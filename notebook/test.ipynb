```python
import pandas as pd
```

```python
df = pd.read_csv("../data/raw/VN_housing_dataset.csv")
```

```python
grouped_df = df.groupby(["Quận","Loại hình nhà ở"])[["Loại hình nhà ở"]].count().unstack("Loại hình nhà ở")
grouped_df = grouped_df.fillna(0)
```

```python
[
        { 
            "name": col[1],
            "type": "bar",
            "data": grouped_df[col].values.tolist(),
        } for col in grouped_df.columns 
]
```

```python
grouped_df.plot(kind="bar", figsize=(20,10))
```
